## Lexcube for Jupyter

Lexcube has recently been released as a Python package for visualizing multidimensional arrays in an interactive way:

```python
import lexcube
```

Let's get the ESDC using `xarray`:

```python
import xarray as xr
```

```python
esdc = xr.open_zarr("http://data.rsc4earth.de/EarthSystemDataCube/v3.0.2/esdc-8d-0.25deg-256x128x128-3.0.2.zarr")
esdc
```

Now, let's visualize a portion of it using `lexcube`!

```python
w = lexcube.Cube3DWidget(esdc["air_temperature_2m"][256:512,256:512,256:512], cmap="thermal")
w
```

Note that you can interact with the cube with this widget!

However, you can also use some sliders:

```python
w.show_sliders()
```

You can also set the limits directly coding them!

```python
w.xlim = (140, 250)
w.ylim = (85, 200)
w.zlim = (0, 140)
```

Furthermore, you can save the current view so you can use them in your publications (remember to cite the [Lexcube paper](https://doi.org/10.1109/MCG.2023.3321989)!)

```python
w.savefig(fname="cube.png", include_ui=True, dpi_scale=3.0)
```

And finally, you can actually use the current visualization as a selection method!

```python
esdc_subset = w.get_current_cube_selection()
esdc_subset
```
